```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

Create basic complex plot.

```python
# Create a figure and axis object
complex_plane = plt.figure()
ax = complex_plane.add_subplot(1, 1, 1)

# Set the limits for the complex plane
ax.set_xlim(-5, 5)  # Real part bounds
ax.set_ylim(-5, 5)  # Imaginary part bounds

# Add axes
ax.axhline(0, color='black', linewidth=0.5)  # Horizontal axis
ax.axvline(0, color='black', linewidth=0.5)  # Vertical axis

# Add title and grid
ax.set_title('Complex Plane')
ax.grid(True)

# Display the plot
plt.show()
```

Define parameters A, B, C in complex plane.

```python
A = -1
alpha = np.pi / 2.5  # TODO: Make this interactive.
B = cmath.exp(1j * alpha)
C = B.conjugate()

complex_points = [A, B, C]
real_parts = []
imag_parts = []

for point in complex_points:
    real_parts.append(point.real)
    imag_parts.append(point.imag)

# Re-create the complex plane to maintain context
complex_plane = plt.figure()
ax = complex_plane.add_subplot(1, 1, 1)
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_title('Complex Plane with Points')
ax.grid(True)

# Scatter plot for the points
ax.scatter(real_parts, imag_parts, color='red')

# Display the plot
plt.show()
```
