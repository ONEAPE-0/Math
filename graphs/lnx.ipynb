```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x values (greater than 0 for ln(x))
x = np.linspace(0.1, 10, 100)  # Avoid 0 to prevent log(0) issues
y = np.log(x)

# Create the plot
plt.plot(x, y, label='ln(x)')
plt.xlabel('x')
plt.ylabel('ln(x)')
plt.title('Graph of ln(x)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
